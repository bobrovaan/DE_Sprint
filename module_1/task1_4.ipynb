# 1. Построение распределений

## Импорт библиотек

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import rayleigh as ray, norm
```

## Распределение Рэлея

```python
q = [0.5, 1.0, 2.0, 3.0, 4.0]
a = np.linspace(0, 10, 1000)
for i in q:
  plt.plot(a, ray(scale = i).pdf(a), label = f'q = {i}')
plt.legend()
plt.grid()
plt.show()
```

## Нормальное распределение

```python
m = [0, 0, 0, -2]
q2 = [0.2, 1.0, 5.0, 0.5]
x = np.linspace(-5, 5, 1000)
for i in range(len(m)):
  plt.plot(x, norm.pdf(x, m[i], q2[i]), label = f'm = {m[i]}, q^2 = {q2[i]}')
plt.legend()
plt.grid()
plt.show()
```

# 2. Построение графика по функции

```python
import numpy as np
from matplotlib import pyplot as plt
from sympy import var, plot_implicit, Eq
```

```python
x, y = var('x y')
arr = [[],[]]
for a, b in plot_implicit(Eq((x ** 2 + y ** 2 - 1) ** 3 - (x ** 2) * (y ** 3), 0), show=False)[0].get_points()[0]:
  arr[0].append(a.mid)
  arr[1].append(b.mid)
plt.scatter(arr[0], arr[1], s = 0.1, c = 'red')
plt.title('(x^2 + y^2 - 1)^3 - x^2 y^3 = 0')
plt.grid()
plt.show()
```

# 3. Анализ зависимостей

## Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Объявление переменных

```python
age = [] # возраст
sex = [] # пол
bmi = [] # индекс массы тела
children = [] # количесво детей
smoke = [] # курит или не курит человек
region = [] # регион
charges = [] # расходы на страховку
```

## Считывание данных с файла

```python
with open("./dataset_home.txt") as file:
    for s in file:
        s = s.split(',')
        age.append(int(s[0].strip()))
        sex.append(s[1].strip())
        bmi.append(float(s[2].strip()))
        children.append(int(s[3].strip()))
        smoke.append(s[4].strip())
        region.append(s[5].strip())
        charges.append(float(s[-1].strip()))

age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoke = np.array(smoke)
region = np.array(region)
charges = np.array(charges)
```

## Предварительная статистика

```python
print(f'Всего обработано {len(age)} строк')
print(f'Минимальный возраст: {age.min()}; Максимальный возраст: {age.max()}; Средний возраст: {round(age.mean(), 1)}')
print(f'Число женщин: {list(sex).count("female")}; Число мужчин: {list(sex).count("male")}')
print(f'Минимальный ИМТ: {bmi.min()}; Максимальный ИМТ: {bmi.max()}; Средний ИМТ: {round(bmi.mean(), 2)}')
print(f'Минимальное число детей: {children.min()}; Максимальное число детей: {children.max()}; Среднее число детей: {round(children.mean())}')
print(f'Число курящих: {list(smoke).count("yes")}; Число некурящих: {list(smoke).count("no")}')
print(f'Минимальные расходы на страховку: {bmi.min()}; Максимальные расходы на страховку: {bmi.max()}; Средние расходы на страховку: {round(bmi.mean(), 2)}')
```
